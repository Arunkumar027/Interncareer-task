```python
# Trend Analysis:

# Identify the most popular categories among the top YouTube streamers
top_streamers = data.sort_values(by="Suscribers", ascending=False).head(10)
top_categories = top_streamers["Categories"].value_counts()

# Visualize the trend of top categories using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title("Most Popular Categories Among Top YouTube Streamers")
plt.xlabel("Category")
plt.ylabel("Number of Streamers")
plt.show()

#"Música y baile" (Music and Dance) is the most popular category among top YouTube streamers, with 3 streamers."Animación, Juguetes" (Animation, Toys) is the second most popular category, with 2 streamers.
#The top YouTube streamers are spread across a diverse range of categories, indicating varied interests and content types among popular streamers.
```

```python
# Calculate the correlation coefficient between subscribers and likes
correlation_subscribers_likes = data["Suscribers"].corr(data["Likes"])

# Calculate the correlation coefficient between subscribers and comments
correlation_subscribers_comments = data["Suscribers"].corr(data["Comments"])

# Print the correlation coefficients
print("Correlation between subscribers and likes:", correlation_subscribers_likes)
print("Correlation between subscribers and comments:", correlation_subscribers_comments)

# Visualize the correlation using scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(data["Suscribers"], data["Likes"])
plt.title("Subscribers vs Likes")
plt.xlabel("Subscribers")
plt.ylabel("Likes")
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data["Suscribers"], data["Comments"])
plt.title("Subscribers vs Comments")
plt.xlabel("Subscribers")
plt.ylabel("Comments")
plt.show()

#Suscribers Vs Likes:  There is a weak positive correlation between the two variables, meaning that channels with more subscribers tend to get more likes, but the relationship is not very strong. This suggests that there are other factors that contribute to the number of likes a video gets besides the number of subscribers a channel has. For example, the quality of the video, the topic of the video, and the marketing strategy used to promote the video can all play a role in the number of likes a video gets.
#Suscribers Vs Comments: This means that as the number of subscribers increases, the number of comments also tends to increase, but the relationship is not very strong. There is a large spread of data points, suggesting that there are many factors besides the number of subscribers that influence the number of comments on a channel.

```
